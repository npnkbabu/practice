```python
#https://queirozf.com/entries/pandas-dataframe-groupby-examples
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
import copy
```

```python
df = pd.read_csv('data/original.csv')
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
#if city is null replace it with No
df_clean = copy.deepcopy(df)
df_clean['clean_city'] = df_clean['City'].fillna('Unknown')
```

```python
df_clean.head(10)
```

```python
df_clean = df_clean[df_clean['JobTitle'].notnull()]
```

```python
#if salary is null replace with avg salray
# clean $
df_clean['clean_salary'] = df['Salary'].map(lambda x : x[1:]).astype(float)
```

```python
mean_sal = df_clean['clean_salary'].mean()
df_clean['new_salary'] = df_clean['clean_salary'].fillna(mean_sal)
df_clean.head(10)
```

```python
#avg salary by gender
df_clean.groupby('gender').mean()['clean_salary']
```

```python
#avg pay of woman and man by jobtitle
df_clean.groupby(['JobTitle','gender']).mean()['new_salary']
```

```python
#city with highes avg salary
df_clean.groupby('City')['new_salary'].mean().to_frame().sort_values(by='new_salary', ascending=False)
```
